```python
df_results =pd.DataFrame(columns=["N","Input_type","SortingAlgo","O2All","O3All","O3unroll","O3only"])
```

```python
df_results
```

```python
O2ALL = 0
O3ALL = 0
O3unroll = 0
O3only = 0

for index, row in df_new.iterrows():
    if (row['Optimization']=='O2ALL') :
        O2ALL = row['RunningTime']
    if (row['Optimization']=='O3ALL') :
        O3ALL = row['RunningTime']
    if (row['Optimization']=='O3unroll') :
        O3unroll = row['RunningTime']
    if (row['Optimization']=='O3only') :
        O3only = row['RunningTime']
        #data = { 'uk':'united kingdom','fr':'france' }
        row_data = {'N' : row['N'],'Input_type' : row['Input_type'], 'SortingAlgo' : row['SortingAlgo'], 'O2All' : O2ALL, 'O3All' : O3ALL, 'O3unroll' : O3unroll, 'O3only' : O3only}
        df_results = df_results.append(row_data,ignore_index=True)

       
```
